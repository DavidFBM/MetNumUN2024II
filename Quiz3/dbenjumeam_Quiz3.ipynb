````
import matplotlib.pyplot as plt
# Cuenta los sobrevivientes y fallecidos
counts = titanic['Survived'].value_counts()
labels = ['Died', 'Survived']
sizes = counts.values
colors = ['#1f77b4', '#ff7f0e']  # Azul y naranja

# Crear la gráfica
plt.figure(figsize=(6, 6))
plt.pie(
    sizes,
    labels=[f'{labels[0]} {sizes[0]}', f'{labels[1]} {sizes[1]}'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
)
plt.title('Died survived proportion')
plt.show()
````

**Analysis**: The graph shows the proportion of people who survived and died in the Titanic disaster. In this case:

*   61.6% of the people died.
*   38.4% of the people survived.

This indicates that most of the passengers did not survive.

````
import matplotlib.pyplot as plt
# Crear el histograma de la distribución de edades
plt.figure(figsize=(8, 6))
plt.hist(titanic['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
````

**Analysis**: The graph shows the age distribution of the Titanic's passengers. Key observations:

*   The majority of passengers were in their 20s and 30s, suggesting that this was a predominant demographic group.
*   There are fewer children and older adults, with a gradual drop in frequencies as age increases.

````
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt
# Crear un diccionario de etiquetas
labels = {
    ('female', 0): 'female\nDied',
    ('female', 1): 'female\nSurvived',
    ('male', 0): 'male\nDied',
    ('male', 1): 'male\nSurvived',
}

# Crear la gráfica de mosaico
plt.figure(figsize=(8, 6))
mosaic(
    titanic,
    ['Sex', 'Survived'],
    labelizer=lambda k: labels.get((k[0], int(k[1])), ''),  # Convertir Survived a entero
    title='Sex vs Survived'
)
plt.show()
````

**Analysis**: The graph shows how the Titanic's passengers were distributed according to gender and survival status:

*   Most men perished (large green box).
*   A significant proportion of women survived compared to men.
*   This could reflect evacuation priorities, where women and children were given preference.

````
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt
# Crear un diccionario de etiquetas
labels = {
    (1, 0): 'Class 1\nDied',
    (1, 1): 'Class 1\nSurvived',
    (2, 0): 'Class 2\nDied',
    (2, 1): 'Class 2\nSurvived',
    (3, 0): 'Class 3\nDied',
    (3, 1): 'Class 3\nSurvived',
}

# Crear la gráfica de mosaico
plt.figure(figsize=(10, 8))
mosaic(
    titanic,
    ['Pclass', 'Survived'],
    labelizer=lambda k: labels.get((int(k[0]), int(k[1])), ''),  # Convertir claves a enteros
    title='Pclass vs Survived'
)
plt.show()
````

**Analysis**: The graph reveals that ticket class was a determining factor in the likelihood of surviving the sinking of the Titanic. First-class passengers had a significant advantage, while third-class passengers faced greater difficulties. This pattern reflects how the social hierarchies of the time influenced even life and death situations.
