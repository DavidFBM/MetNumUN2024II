# **Modificaciones**




*   **Cambio del Modelo de Machine Learning** : La regresión lineal se cambió por un árbol de desiciones. La regresión lineal asume una relación lineal entre variables, lo cual puede ser una limitación. Un árbol de decisiones permite capturar relaciones no lineales en los datos, mejorando potencialmente la precisión del modelo.

*   **El mercado y activos** : Se cambió el mercado de NASDAQ a S&P 500. Se selecionaron todos los activos del mercado S&P 500.

*   **Las variables** : Se agregaron 2 nuevas variables:  ATR (Average True Range), Mide la volatilidad del activo considerando la diferencia entre el precio máximo y mínimo, ajustado con el precio de cierre anterior, un ATR alto indica alta volatilidad, mientras que un ATR bajo sugiere un mercado más estable. OBV (On-Balance Volume), analiza la relación entre el volumen y los cambios de precio para medir la presión de compra o venta, si el precio sube con un OBV creciente, sugiere una tendencia alcista fuerte. Si el precio sube pero el OBV baja, indica una posible debilidad en la tendencia.

*   **Cambio en la Clasificación de los Targets** : Se amplió la clasificación para incluir una categoría neutral: 1 si el precio sube más del 0.5%, 0 si el cambio está dentro de ±0.5% y -1 si el precio baja más del 0.5%
