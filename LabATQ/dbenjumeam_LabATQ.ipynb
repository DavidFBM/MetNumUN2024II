# **Modificaciones**




*   **Cambio del Modelo de Machine Learning** : La regresión lineal se cambió por un árbol de desiciones. La regresión lineal asume una relación lineal entre variables, lo cual puede ser una limitación. Un árbol de decisiones permite capturar relaciones no lineales en los datos, mejorando potencialmente la precisión del modelo.

*   **El mercado y activos** : Se cambió el mercado de NASDAQ a S&P 500. Se selecionaron todos los activos del mercado S&P 500.

*   **Las variables** : Se agregaron 2 nuevas variables:  ATR (Average True Range), Mide la volatilidad del activo considerando la diferencia entre el precio máximo y mínimo, ajustado con el precio de cierre anterior, un ATR alto indica alta volatilidad, mientras que un ATR bajo sugiere un mercado más estable. OBV (On-Balance Volume), analiza la relación entre el volumen y los cambios de precio para medir la presión de compra o venta, si el precio sube con un OBV creciente, sugiere una tendencia alcista fuerte. Si el precio sube pero el OBV baja, indica una posible debilidad en la tendencia. Adicionalmente se agregaron tres observaciones previas (lags) para cada variable, lo que permite capturar tendencias y patrones históricos. Esto se hizo desplazando cada serie de datos en el tiempo y agregándolas como nuevas características con sufijos _lag1, _lag2 y _lag3.
